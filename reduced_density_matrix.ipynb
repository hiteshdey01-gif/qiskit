```python
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
from math import sqrt
```

```python
qc = QuantumCircuit(2)
qc.h(1)
qc.cx(1,0)
qc.draw('mpl',reverse_bits=True)
```

```python
# Total density matrix of the system
rhoAB = qi.DensityMatrix(qc)
rhoAB.draw('latex')
```

```python
# Reduced density matrix for B
rhoB = qi.partial_trace(rhoAB,[1])
rhoB.draw('latex')
```

```python
# Reduced density matrix for A
rhoA = qi.partial_trace(rhoAB,[0])
rhoA.draw('latex')
```

```python
# another state
psi2= 1/sqrt(3)*qi.Statevector([0,1,1,0,1,0,0,0])
psi2.draw('latex')
```

```python
rho2 = qi.DensityMatrix(psi2)
rho2.draw('latex')
```

```python
traceB = qi.partial_trace(rho2,[0,1])
traceB.draw('latex')
```
