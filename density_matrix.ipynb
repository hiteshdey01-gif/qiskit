```python
import numpy as np
```

```python
# defining the bell state
psi = np.array([1,0,0,1])/np.sqrt(2)
print("State vector |psi> =", psi)
```

```python
# Density Matrix of the state |psi><psi|
rho = np.outer(psi, psi.conj())
print("Density matrix rho =\n", rho)
```

```python
print("Trace of rho =", np.trace(rho))
```

### Trace of the density matrix should be equal to 1

```python
rho_tensor =rho.reshape(2,2,2,2)
print("shape of rho tensor =", rho_tensor.shape)
print(rho_tensor[0,0,0,0])
```

### calulating the partial trace

```python
# Initialize reduced density matrix
rho_A = np.zeros((2,2), dtype=complex)

# Perform partial trace over B
for a in range(2):
    for c in range(2):
        for b in range(2):
            rho_A[a, c] += rho_tensor[a, b, c, b]

print("Reduced density matrix rho_A =")
print(rho_A)
```

```python
rho_A_fast = np.trace(rho_tensor, axis1=1, axis2=3)

print("Reduced density matrix (fast method) =")
print(rho_A_fast)
```
