# Measurements

```python
import numpy as np
```

```python
def measure(state):
    probabilities = np.abs(state) ** 2
    outcome = np.random.choice(len(state), p=probabilities)  # random outcomes
    return outcome
```

```python
psi = np.array([1/np.sqrt(2), 1/np.sqrt(2)])

results = [measure(psi) for _ in range(20)]
print(results)
```

```python
N = 20
print("0:", results.count(0))
print("1:", results.count(1))
```

```python
psi = np.array([1/np.sqrt(3), 1/np.sqrt(3), 1/np.sqrt(3)])
results = [measure(psi) for _ in range(30000)]
print(results.count(0))
print(results.count(1))
print(results.count(2))
```
