# INNER PRODUCT

```python
import numpy as np
```

```python
ket0 = np.array([1,0],dtype=complex)
ket1 = np.array([0,1],dtype=complex)
print(ket0)
print(ket1)
```

```python
# General state
psi = 1/np.sqrt(2)*ket0 + 1/np.sqrt(2)*ket1
print(psi)
```

```python
inner_product = np.vdot(psi,psi)
print(inner_product)
```

```python
# Check if the state is normalized
def is_normalized(state):
    inner_product = np.vdot(state,state)
    return True if abs(inner_product - 1) < 1e-10 else False
```

```python
# another method to check normalization
def is_normalized(state):
    return np.isclose(np.vdot(state,state),1)   # np.isclose checks if the two values are close to each other within a certain tolerance
```

# Introduction of global phase

```python
psi1 = np.array([1/np.sqrt(2), 1/np.sqrt(2)],dtype=complex)
psi2 = np.array([1/np.sqrt(2), -1/np.sqrt(2)],dtype=complex)
psi3 = -psi1

print("psi1:", psi1)
print("psi2:", psi2)
print("psi3:", psi3)
```

```python
# compare probability amplitudes

def probabilities(state):
    return np.abs(state)**2
print("Probabilities for psi1:", probabilities(psi1))
print("Probabilities for psi2:", probabilities(psi2))   
print("Probabilities for psi3:", probabilities(psi3))
```
