```python
# AND OPERATION
0 & 0
0 & 1
1 & 0
#1 & 1
```

```python
1010 & 1111 
```

Using AND  operation between two larger binary number does not work. So we use 0b prefix, and then convert into binary. 

```python
0b1010 & 0b1111 
```

```python
bin(0b1010 & 0b1111)
```

```python
# OR OPERATION
0 | 0
0 | 1
1 | 0
1 | 1
```

```python
bin(0b1010 | 0b1110)
```

```python
bin(0b1010 ^ 0b1110) # XOR OPERATION
```

Here we expected 0100 but python removes the 0 at the start

Now to get a NOT operation, as python does not have an inbuilt NOT operation so we do this using XOR operation

```python
0 ^ 1
```

When we do 0 XOR 1 we get 1, so we can think of this 0 as our input and 1 as our operator. So when we want to use the NOT operation we apply XOR between input and a string of 1's.

```python
bin(0b0010 ^ 0b1111) # NOT OPERATION
```
