# Hadamard gate

```python
import numpy as np
```

```python
# Hadamard gate
H = (1/np.sqrt(2))* np.array([[1,1],
                              [1,-1]])
print(H)
```

```python
psi1 = np.array([1/np.sqrt(2), 1/np.sqrt(2)],dtype=complex)
psi2 = np.array([1/np.sqrt(2), -1/np.sqrt(2)],dtype=complex)
```

```python
# Applying the Hadamard gate 
#psi1_after_H = np.dot(H, psi1)
#psi2_after_H = np.dot(H, psi2)  
psi1_after_H = H @ psi1     # @ is an alternative to np.dot for matrix multiplication
psi2_after_H = H @ psi2

print("psi1 after applying H:", psi1_after_H)
print("psi2 after applying H:", psi2_after_H)
```

```python
def probabilities(state):
    return np.abs(state)**2

print("Probabilities for psi1 after H:", probabilities(psi1_after_H))
print("Probabilities for psi2 after H:", probabilities(psi2_after_H))
```

## Final Understanding

Measurement alone doesn’t reveal phase\
