```python
# So we use
b1=bin(5)
b2=bin(2)
x1=int(b1,2) # turning them back into integer. 2 represents base here
print(x1)
x2=int(b2,2)
print(x2)
x1+x2
```

```python
for i in range (8): # printing numbers from 0 to 7 in decimal 
    print(i)
```

```python
# for binary
for i in range(8):
    b=bin(i)
    print(b)
```

```python
# for removing the 0b
for i in range(8):
    b=bin(i)[2:]  # 2 indicates that we need the output after 2 character
    print(b)
```

```python
# for pdding
for i in range(8):
    b=bin(i)[2:].zfill(3)
    print(b)
```

```python
# An easy way to do all this is using numpy
import numpy as np
np.binary_repr(13,4) # 13 in the number and 4 represents number of bits
```

```python
for i in range(8):
    b=np.binary_repr(i,4)
    print(b)
    
    
```
