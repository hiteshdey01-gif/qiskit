## Pauli- X Gate 

```python
import numpy as np
```

```python
ket0 = np.array([1,0], dtype=complex)
ket1 = np.array([0,1], dtype=complex)
```

```python
# Pauli-X gate
X = np.array([[0, 1],
               [1, 0]], dtype=complex)
```

```python
# apply x gate
result0 = X @ ket0
result1 = X @ ket1
print("X|0> =", result0)
print("X|1> =", result1)
```

```python
psi = np.array([1/np.sqrt(2), 1/np.sqrt(2)])

psi_new = X @ psi

print("X|psi> =", psi_new)
```

```python
psi = np.array([1/np.sqrt(2), -1/np.sqrt(2)], dtype=complex)

psi_new = X @ psi

print("X|psi> =", psi_new)
```

## pauli z gate


```python
Z = np.array([[1,0],
             [0,-1]], dtype=complex)

psi = np.array([1/np.sqrt(2), 1/np.sqrt(2)], dtype=complex)
psi_minus = np.array([1/np.sqrt(2), -1/np.sqrt(2)], dtype=complex)
psi_new = Z @ psi
psi_minus_new = Z @ psi_minus
print("Z|psi> =", psi_new)
print("Z|psi-> =", psi_minus_new)

```

Z flips the relative phase
(+) ↔ (−)

```python
# Apply Z gate twice gives back the original state

psi = np.array([1/np.sqrt(2), 1/np.sqrt(2)], dtype=complex)
psi_minus = np.array([1/np.sqrt(2), -1/np.sqrt(2)], dtype=complex)

psi_new = Z @ (Z @ psi)
psi_minus_new = Z @ (Z @ psi_minus)
print("Z(Z|psi>) =", psi_new)
print("Z(Z|psi->) =", psi_minus_new)
```

### Z is a reversible operation
quantum gates must be reversible (unitary)

# Hadamard Gate

```python
H = (1/np.sqrt(2)) * np.array([[1, 1],
                              [1, -1]], dtype=complex)

ket0 = np.array([1,0], dtype=complex)
ket1 = np.array([0,1], dtype=complex)

result0 = H @ ket0
result1 = H @ ket1
print("H|0> =", result0)
print("H|1> =", result1)
```

```python
# Applying  twice gives back the original state
psi = np.array([1, 0], dtype=complex)  # |0>
psi_new = H @ (H @ psi)
print(psi_new)
```

## Hadamard is special because:

 It connects X and Z 

In fact: HZH = X \
         HXH = Z

This  means:

Z in one basis = X in another basis

```python
left = H @ Z @ H
print("H Z H =", left)

print("X =", X)
```

```python
left2 = H @ X @ H
print("H X H =", left2)

print("Z =", Z)
```
